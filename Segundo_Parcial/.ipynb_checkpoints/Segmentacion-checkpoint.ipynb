```python
# --- SegmentaciÃ³n 1 --- #
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carga la imagen
ruta_imagen = "C:/Users/luisl/OneDrive/Escritorio/PDI/Estancia Investigacion/Recursos/moneda.png"
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Binariza la imagen
_, imagenBinarizada = cv2.threshold(imagen, 120, 255, cv2.THRESH_BINARY)

# Invertir la imagen binarizada
imagen_binarizada = cv2.bitwise_not(imagenBinarizada)
alto, ancho = imagen_binarizada.shape

def asignar_etiqueta(vecindad, etiquetas):
    vecindad = vecindad.flatten()
    vecindad = vecindad[vecindad > 1]
    if len(vecindad) > 0:
        return np.min(vecindad)
    else:
        return None


etiquetas = np.zeros_like(imagen_binarizada, dtype=np.uint16)

etiqueta_actual = 1

for y in range(alto):
    for x in range(ancho):
        if imagen_binarizada[y, x] != 0:
            vecindad = etiquetas[max(0, y-1):min(alto, y+2), max(0, x-1):min(ancho, x+2)]
            etiqueta = asignar_etiqueta(vecindad, etiquetas)
            if etiqueta is None:
                etiqueta_actual += 1
                etiquetas[y, x] = etiqueta_actual
            else:
                etiquetas[y, x] = etiqueta

# Crear una imagen con colores aleatorios para cada objeto identificado
colores = np.random.randint(0, 255, (etiqueta_actual+1, 3), dtype=np.uint8)
objetos_coloreados = np.zeros((alto, ancho, 3), dtype=np.uint8)

for etiqueta in np.unique(etiquetas):
    if etiqueta != 0:
        objetos_coloreados[etiquetas == etiqueta] = colores[etiqueta]


# Resultados
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(objetos_coloreados, cmap='gray')
plt.title("Objetos identificados")
plt.axis('off')

plt.show()
```
