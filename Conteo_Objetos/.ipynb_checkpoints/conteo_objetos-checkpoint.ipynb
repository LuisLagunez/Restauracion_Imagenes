```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ruta_imagen = "C:/Users/luisl/OneDrive/Documentos/GitHub/Restauracion_Imagenes/Recursos/bacterias.jpg"

# Cargar la imagen y redimensionarla
imagen = cv2.imread(ruta_imagen)
imagen = cv2.resize(imagen, (int(imagen.shape[1]/10), int(imagen.shape[0]/10)))

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar operaciones morfológicas
kernel = np.ones((5, 5), dtype=np.uint8)
erosion = cv2.erode(imagen_gris, kernel, iterations=3)
dilatacion = cv2.dilate(erosion, kernel, iterations=8)
gauss = cv2.GaussianBlur(dilatacion, (5, 5), 0)
canny = cv2.Canny(gauss, 100, 200)

# Encontrar contornos en la imagen Canny
contornos, _ = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujar contornos en la imagen original y contar objetos
imagen_contornos = imagen.copy()
cv2.drawContours(imagen_contornos, contornos, -1, (0, 0, 255), 2)
numero_objetos = len(contornos)

# Mostrar número de objetos en la imagen
texto = 'Objetos encontrados: ' + str(numero_objetos)
cv2.putText(imagen_contornos, texto, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

# Mostrar resultados
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen_contornos, cv2.COLOR_BGR2RGB))
plt.title("Imagen con contornos")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(canny, cmap='gray')
plt.title("Detección de bordes (Canny)")
plt.axis('off')

plt.show()
```
